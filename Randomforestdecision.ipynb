```python
# Initial imports.
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Loading data
file_path = Path("../Final_Project/All_Fighters_aligned.csv")
df_fighters = pd.read_csv(file_path)
df_fighters.head()
```

```python
df_fighters = df_fighters.dropna()
df_fighters
```

```python
# Define the features set.
X = df_fighters.copy()
X = X.drop("Results", axis=1)
X.head()
```

```python
# Define the target set.
y = df_fighters["Results"].values
y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Determine the shape of our training and testing sets.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=1) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test)
predictions
```

```python
# Calculating the confusion matrix.
conmat = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
conmat_df = pd.DataFrame(
    conmat, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

conmat_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
# Displaying results
print("Confusion Matrix")
display(conmat_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
